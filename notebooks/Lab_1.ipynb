```python
import jpype
import matplotlib.pyplot as plt
```

```python
jpype.startJVM(classpath=[".."])
```

```python
Fibonacci = jpype.JClass("utils.fibonacci_algoritms.Fibonacci")

rec_results = []
iter_results = []

for n in range(50):
    rec_result = Fibonacci.fibonacciRec(n)
    rec_results.append((rec_result.n, rec_result.result, rec_result.iterations, rec_result.timeMs))
    
    iter_result = Fibonacci.fibonacciIter(n)
    iter_results.append((iter_result.n, iter_result.result, iter_result.iterations, iter_result.timeMs))
```

```python
n_values = [x[0] for x in rec_results]

rec_result = [x[1] for x in rec_results]
iter_result = [x[1] for x in iter_results]

rec_iteration = [x[2] for x in rec_results]
iter_iteration = [x[2] for x in iter_results]

rec_time = [x[3] for x in rec_results]
iter_time = [x[3] for x in iter_results]
```

```python
import pandas as pd

results_dataframe = pd.DataFrame({
    "Рекурсивный метод Результат": rec_result,
    "Итеративный метод Результат": iter_result,
    "Рекурсивный метод время (мс)": rec_time,
    "Итеративный метод время (мс)": iter_time,
    "Рекурсивный метод Число итераций": rec_iteration,
    "Итеративный метод Число итераций": iter_iteration,

}, index=n_values)

results_dataframe
```

```python
fig, axs = plt.subplots(2, 1, figsize=(16, 20)) 

axs[0].plot(n_values[10:35], rec_time[10:35], label='Рекурсивный метод', color='red')
axs[0].plot(n_values[10:35], iter_time[10:35], label='Итеративный метод', color='blue')

axs[0].set_xlabel('n')  
axs[0].set_ylabel('Время выполнения (мс)')
axs[0].set_title('Сравнение времени выполнения')

axs[0].legend()
axs[0].grid(True)


axs[1].plot(n_values[10:35], rec_iteration[10:35], label='Рекурсивный метод', color='red')
axs[1].plot(n_values[10:35], iter_iteration[10:35], label='Итеративный метод', color='blue')

axs[1].set_xlabel('n')  
axs[1].set_ylabel('Число итераций (для итеративного количество итераций цикла, для рекурсивного количество вызовов функции)')
axs[1].set_title('Сравнение числа итераций')

axs[1].legend()
axs[1].grid(True)


plt.show()
```

Сложность рекурсивного метода:
- По времени: O(2^n)
- По памяти: O(n)

Сложность итеративного метода:
- По времени: O(n)
- по памяти: O(1)
